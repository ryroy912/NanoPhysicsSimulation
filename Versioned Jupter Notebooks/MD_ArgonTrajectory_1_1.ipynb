**Visualization of one single atom from the collection of atoms over the entire timeseries**

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, IntSlider
from mpl_toolkits.mplot3d import Axes3D

# Extract all 10,000 positions of the first atom from df2
first_atom_df = df2[df2["Atom ID"] == 1].copy()  # Ensure correct Atom ID
first_atom_positions = first_atom_df[["X (nm)", "Y (nm)", "Z (nm)"]].values * 10  # Convert nm to Å

# Function to update the 3D plot
def update(frame_index):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection="3d")

    # Plot full trajectory up to current frame
    ax.plot(first_atom_positions[:frame_index, 0],
            first_atom_positions[:frame_index, 1],
            first_atom_positions[:frame_index, 2], "b", alpha=0.5)

    # Highlight the current position
    ax.scatter(first_atom_positions[frame_index, 0],
               first_atom_positions[frame_index, 1],
               first_atom_positions[frame_index, 2],
               c="r", s=100, label="Current Position")

    ax.set_xlabel("X (Å)")
    ax.set_ylabel("Y (Å)")
    ax.set_zlabel("Z (Å)")
    ax.set_title(f"Argon Atom Trajectory - Time Step {frame_index * 100}")

    plt.legend()
    plt.show()

# Create interactive slider for visualization
interactive_plot = interactive(update, frame_index=IntSlider(min=0, max=len(first_atom_positions)-1, step=1000, value=0))
interactive_plot
```

```python
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader

class ArgonTrajectoryDataset(Dataset):
    def __init__(self, df2, input_steps=1000, output_steps=100):
        """
        Create sliding windows from the Argon trajectory dataset.
        """
        self.input_steps = input_steps
        self.output_steps = output_steps

        # Extract atomic positions and reshape into (timesteps, 512, 3)
        atom_positions = df2.pivot(index="Time (ps)", columns="Atom ID", values=["X (nm)", "Y (nm)", "Z (nm)"])
        atom_positions = atom_positions.values.reshape(len(df2["Time (ps)"].unique()), 512, 3)

        # Normalize the data
        self.data = (atom_positions - np.mean(atom_positions)) / np.std(atom_positions)

        # Create input-output pairs using a sliding window
        self.inputs = []
        self.outputs = []
        for i in range(len(self.data) - input_steps - output_steps):
            self.inputs.append(self.data[i : i + input_steps])
            self.outputs.append(self.data[i + input_steps : i + input_steps + output_steps])

        self.inputs = np.array(self.inputs)  # Shape: (num_samples, 100, 512, 3)
        self.outputs = np.array(self.outputs)  # Shape: (num_samples, 10, 512, 3)

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return torch.tensor(self.inputs[idx], dtype=torch.float32), torch.tensor(self.outputs[idx], dtype=torch.float32)

# Create dataset and dataloader
dataset = ArgonTrajectoryDataset(df2)
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
```
